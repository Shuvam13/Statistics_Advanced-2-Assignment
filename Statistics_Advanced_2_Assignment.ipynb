1. What is hypothesis testing in statistics?
   - Hypothesis testing is a statistical method used to determine whether there is enough evidence in a sample of data to infer that a certain condition is true for the entire population.
   - It's a way to make decisions or judgments about parameters of a population based on sample data.

2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?
   - The null hypothesis is a statement that there is no significant difference between two populations or variables. It is the default assumption that is tested against.
   - Here's a simple way to think about the difference:
     - Null Hypothesis: Represents the status quo or no effect.
     - Alternative Hypothesis: Represents what you are trying to find evidence for (an effect or a difference).

3. Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.
   - The significance level, often denoted by alpha, is the probability of rejecting the null hypothesis when it is actually true.
   - It is a threshold that you set before conducting the hypothesis test.
   - It represents the maximum risk of committing a Type I error (false positive) that you are willing to accept.
   - Here's its role in deciding the outcome of a test:
     - You choose a significance level (commonly 0.05, 0.01, or 0.10) before you start the test. A significance level of 0.05 means you are willing to accept a 5% chance of incorrectly rejecting the null hypothesis.
     - After performing the statistical test, you obtain a p-value.
     - If the p-value is less than or equal to the significance level, you reject the null hypothesis.
     - If the p-value is greater than the significance level, you fail to reject the null hypothesis.

4. What are Type I and Type II errors? Give examples of each.
   - A Type I error occurs when you reject the null hypothesis when it is actually true.
   - EXAMPLE: The study concludes that the drug lowers blood pressure, but in reality, it has no effect.
   - A Type II error occurs when you fail to reject the null hypothesis when it is actually false.
   - EXAMPLE: The study concludes that the drug has no effect on blood pressure, but in reality, the drug does effectively lower blood pressure.
   
5. What is the difference between a Z-test and a T-test? Explain when to use
each.
   - The Z-test and T-test are both statistical hypothesis tests used to determine if there is a significant difference between the means of two groups or between a sample mean and a hypothesized population mean.
   - The main difference between them lies in the information we have about the population standard deviation and the sample size.
   - USES OF Z-TEST:
     - A Z-test is used when you know the population standard deviation and the sample size is large.
     - It assumes that the data is normally distributed.
     - The Z-statistic is calculated using the population standard deviation.

   - USES OF T-TEST:
     - A T-test is used when you do not know the population standard deviation and you have to estimate it using the sample standard deviation.
     - It assumes that the data is approximately normally distributed .
     - The T-statistic is calculated using the sample standard deviation and the degrees of freedom.

6. Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.

   Hint: Generate random number using random function.

```python
import numpy as np
import matplotlib.pyplot as plt

n = 15
p = 0.5
size = 1000

binomial_data = np.random.binomial(n, p, size)

plt.hist(binomial_data, bins=np.arange(-0.5, n+1.5, 1), edgecolor='black', density=True)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution (n=15, p=0.5)')
plt.xticks(range(n+1))
plt.grid(axis='y', alpha=0.75)
plt.show()
```

7. Implement hypothesis testing using Z-statistics for a sample dataset in
Python.

   Show the Python code and interpret the results.

   sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
   50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
   50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
   50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

```python
import numpy as np
from statsmodels.stats.weightstats import ztest

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mu_0 = 50
alpha = 0.05
z_statistic, p_value = ztest(sample_data, value=mu_0)

print(f"Z-statistic: {z_statistic}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("Reject the null hypothesis: There is enough evidence to suggest the sample mean is significantly different from the hypothesized population mean.")
else:
    print("Fail to reject the null hypothesis: There is not enough evidence to suggest the sample mean is significantly different from the hypothesized population mean.")
```

8. Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(42)
mu = 10
sigma = 2
sample_size = 100

simulated_data = np.random.normal(mu, sigma, sample_size)

sample_mean = np.mean(simulated_data)
sample_std = np.std(simulated_data, ddof=1)
confidence_level = 0.95
alpha = 1 - confidence_level

z_critical = norm.ppf(1 - alpha/2)
margin_of_error = z_critical * (sample_std / np.sqrt(sample_size))

confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Sample Mean: {sample_mean}")
print(f"Sample Standard Deviation: {sample_std}")
print(f"Z-critical value: {z_critical}")
print(f"Margin of Error: {margin_of_error}")
print(f"95% Confidence Interval for the Mean: {confidence_interval}")

plt.figure(figsize=(8, 6))
plt.hist(simulated_data, bins=15, density=True, alpha=0.7, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Simulated Data from Normal Distribution')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

9. Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram.

   Explain what the Z-scores represent in terms of standard deviations from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):

  mean = np.mean(data)
  std_dev = np.std(data)
  z_scores = (data - mean) / std_dev
  return z_scores

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

z_scores_data = calculate_z_scores(sample_data)

print("Original Data (first 10):", sample_data[:10])
print("Z-scores (first 10):", z_scores_data[:10])

plt.figure(figsize=(8, 6))
plt.hist(z_scores_data, bins=15, density=True, alpha=0.7, color='lightgreen', edgecolor='black')
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.title('Histogram of Z-scores')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

**Explanation of Z-scores:**

A Z-score measures how many standard deviations away from the mean a particular data point is.

*   A Z-score of 0 means the data point is exactly at the mean.
*   A positive Z-score means the data point is above the mean. For example, a Z-score of 1 means the data point is one standard deviation above the mean.
*   A negative Z-score means the data point is below the mean. For example, a Z-score of -1 means the data point is one standard deviation below the mean.

Z-scores are useful for standardizing data, which allows you to compare data points from different datasets with different means and standard deviations. They also play a key role in many statistical tests and analyses, including hypothesis testing.
